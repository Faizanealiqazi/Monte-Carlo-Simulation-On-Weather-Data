```python
columns_to_not_update = ['date']

def fill_with_average(df, columns_to_exclude=[]):
    for column in df.columns.difference(columns_to_exclude):
        if df[column].dtype.kind in 'biufc':
            for i in df[df[column].isna()].index:

                prev_index = df[column].iloc[:i].last_valid_index()
                next_index = df[column].iloc[i+1:].first_valid_index()

                if pd.notnull(prev_index) and pd.notnull(next_index):
                    df.at[i, column] = (df.at[prev_index, column] + df.at[next_index, column]) / 2

    return df
df1 = pd.DataFrame({
    'Value': [10, np.nan, np.nan, np.nan, 20] #=> 10, 15, 17.5, 18.75, 20
})
df = fill_with_average(df)
df
```

```python
for col in df.columns:
    if col != 'date':
        df[col] = pd.to_numeric(df[col], errors='coerce')

print(df.dtypes)
```

```python
def MonteCarloSimulation(df):
    df['date'] = pd.to_datetime(df['date'], format='%Y-%d-%m')

    daily_changes = df.select_dtypes(include=['float64']).diff()
    daily_changes = daily_changes.dropna()

    resultant_data = {}

    prediction_dates = pd.date_range(start='2023-11-06', end='2023-12-06')

    last_known_values = df.iloc[-1][daily_changes.columns].to_dict()

    for date in prediction_dates:
        simulated_day = last_known_values.copy()
        for col in daily_changes.columns:
            simulated_change = np.random.normal(daily_changes[col].mean(), daily_changes[col].std())
            simulated_day[col] = max(0, simulated_day[col] + simulated_change)

        resultant_data[date] = simulated_day

    result = pd.DataFrame.from_dict(resultant_data, orient='index')
    return result
```

```python
final_df = MonteCarloSimulation(df)
final_df
```
